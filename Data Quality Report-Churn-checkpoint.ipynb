## Generating Report for Categorical Variables 
    These variables take on values that are names or labels

```python
# Defining Empty Containers
null_perc_dict = {} 
count_dict = {} 
cardinality_dict = {} 
mode_name = [] 
mode_value = [] 
mode_perc = [] 
mode_dist_perc = []
mode2_name = [] 
mode2_value = [] 
mode2_perc = [] 
mode2_dist_perc = []
```

```python
# Calculating required values
for i in range(len(cat_var)): 
    # Calculating null percentage value
    value = round(df[cat_var[i]].isnull().sum()*100 / len(df),2) 
    null_perc_dict[cat_var[i]] = value  
    # Calculating Distinct Count
    count_dict[cat_var[i]] = df[cat_var[i]].count() 
    # Calculating Cardinality
    cardinality_dict[cat_var[i]] = df[cat_var[i]].nunique() 
    # Calculating Mode Properties
    mode_name.append(df[cat_var[i]].value_counts().index.tolist()[0]) 
    mode_value.append(df[cat_var[i]].value_counts().tolist()[0]) 
    mode_perc.append(round((df[cat_var[i]].value_counts().tolist()[0] / len(df))*100, 2)) 
    mode_dist_perc.append(round((df[cat_var[i]].value_counts().tolist()[0] / df[cat_var[i]].count())*100, 2)) 
    # Calculating 2nd Mode Properties(if it exists)
    if(df[cat_var[i]].nunique() > 1) : 
        mode2_name.append(df[cat_var[i]].value_counts().index.tolist()[1]) 
        mode2_value.append(df[cat_var[i]].value_counts().tolist()[1]) 
        mode2_perc.append(round((df[cat_var[i]].value_counts().tolist()[1] / len(df))*100, 2)) 
        mode2_dist_perc.append(round((df[cat_var[i]].value_counts().tolist()[1] / df[cat_var[i]].count())*100, 2))
    else : 
        mode2_name.append('N.A.') 
        mode2_value.append('N.A.') 
        mode2_perc.append('N.A.') 
        mode2_dist_perc.append('N.A.')
```

```python
# Creating a dataframe from the calculated values
cat_df = pd.DataFrame([count_dict, null_perc_dict, cardinality_dict]) 
cat_df = cat_df.transpose() 
cat_df.columns = ['Distinct Count', 'Missing%', 'Cardinality'] 

cat_df['Mode'] = mode_name 
cat_df['Mode Freq.'] = mode_value 
cat_df['Mode%'] = mode_perc 
cat_df['Mode Dist%'] = mode_dist_perc 

cat_df['2nd Mode'] = mode2_name 
cat_df['2nd Mode Freq.'] = mode2_value 
cat_df['2nd Mode%'] = mode2_perc 
cat_df['2nd Mode Dist%'] = mode2_dist_perc 
```

```python
# defining data types 
boolean_features = df.columns[df.dtypes == 'bool'].tolist() 
datetime_features = [] 
alphanumeric_features = [] 

for i in df.columns[df.dtypes == 'object'].tolist(): 
    if(False not in df[i].str.isalnum().tolist() and False in df[i].str.isalpha().tolist()): 
        alphanumeric_features.append(i)  
    else: 
        try: 
            df[i] = pd.to_datetime(df[i]) 
            datetime_features.append(i) 
        except: 
            continue 
# Adding 'data type' column in the dataframe
cat_df['DataType'] = 'abc'
for i in range(len(cat_df)): 
    if(cat_df.index[i] in datetime_features): 
        cat_df['DataType'][i] = 'DateTime' 
    elif(cat_df.index[i] in alphanumeric_features): 
        cat_df['DataType'][i] = 'AlphaNumeric' 
    elif(cat_df.index[i] in boolean_features): 
        cat_df['DataType'][i] = 'Boolean'
    else: 
        cat_df['DataType'][i] = 'Categorical'
```

```python
# Adding 'Data Source' column in the dataframe 
if error == '' :  
    cat_features = cat_df.index.to_list() 
    filtered_source = data_source_df[data_source_df['Features'].isin(cat_features)] 
    cat_df = pd.merge(cat_df, filtered_source, left_on=cat_df.index, right_on='Features') 
    cat_df = cat_df.set_index('Features') 
    cat_df['TotalRows'] = len(df)
    cat_df.to_csv('Categorical_Data_Quality_Report.csv', index=True)
else : 
    cat_df['DataSource'] = "Default Data Source"   
    cat_df['Features'] = cat_df.index 
    cat_df = cat_df[['Features', 'Distinct Count', 'Missing%', 'Cardinality', 'Mode', 'Mode Freq.',
               'Mode%', 'Mode Dist%', '2nd Mode', '2nd Mode Freq.', '2nd Mode%',
               '2nd Mode Dist%', 'DataType', 'DataSource'] ] 
    cat_df['TotalRows'] = len(df)
    cat_df.to_csv('Categorical_Data_Quality_Report.csv', index=False)
    
# Adding Total Rows column 
#cat_df['TotalRows'] = len(df)
```

## Generating Report for Continious Variables 
    These variables are numerical. They represent a measurable quantity.

```python
# Defining Empty Containers
null_perc_dict = {} 
count_dict = {} 
cardinality_dict = {} 
min_dict = {} 
max_dict = {} 
mean_dict = {} 
median_dict = {} 
quartile1_dict = {} 
quartile3_dict = {} 
std_dict = {}
```

```python
# Calculating Required Values
for i in range(len(cont_var)): 
    # Calculating null percentage value
    value = round(df[cont_var[i]].isnull().sum()*100 / len(df),2) 
    null_perc_dict[cont_var[i]] = value 
    # Calculating Distinct Count
    count_dict[cont_var[i]] = df[cont_var[i]].count() 
    # Calculating Cardinality
    cardinality_dict[cont_var[i]] = df[cont_var[i]].nunique() 
    # Calculating the Minimum value
    min_dict[cont_var[i]] = round(df[cont_var[i]].min(),2) 
    # Calculating the Maximum value
    max_dict[cont_var[i]] = round(df[cont_var[i]].max(),2) 
    # Calculating the Mean
    mean_dict[cont_var[i]] = round(df[cont_var[i]].mean(),2) 
    # Calculating the Median
    median_dict[cont_var[i]] = round(df[cont_var[i]].median(),2) 
    # Calculating the 1st Quartile
    quartile1_dict[cont_var[i]] = round(df[cont_var[i]].quantile([0.25]).tolist()[0],2) 
    # Calculating the 3rd Quartile
    quartile3_dict[cont_var[i]] = round(df[cont_var[i]].quantile([0.75]).tolist()[0],2) 
    # Calculating the Standard Deviation
    std_dict[cont_var[i]] = round(df[cont_var[i]].std(),2)
```

```python
# Creating a Dataframe from the calculated values
cont_df = pd.DataFrame([count_dict, null_perc_dict, cardinality_dict, min_dict, 
                        quartile1_dict, mean_dict, median_dict, quartile3_dict, max_dict, std_dict]) 
cont_df = cont_df.transpose() 
cont_df.columns = ['Count', 'Missing%', 'Cardinality', 'Min', '1st Qrt', 'Mean', 'Median', '3rd Qrt', 'Max', 'Std Dev.']
```

```python
# Adding 'Data Source' column in the dataframe 
if error == '' :
    cont_features = cont_df.index.to_list() 
    filtered_source = data_source_df[data_source_df['Features'].isin(cont_features)] 
    cont_df = pd.merge(cont_df, filtered_source, left_on=cont_df.index, right_on='Features') 
    cont_df = cont_df.set_index('Features') 
    cont_df['TotalRows'] = len(df)
    cont_df.to_csv('Continious_Data_Quality_Report.csv',index=True)
else : 
    cont_df['Data Source'] = "Default Data Source" 
    cont_df['Features'] = cont_df.index 
    cont_df = cont_df[['Features', 'Count', 'Missing%', 'Cardinality', 'Min', '1st Qrt', 'Mean', 'Median',
                   '3rd Qrt', 'Max', 'Std Dev.', 'Data Source']] 
    cont_df['TotalRows'] = len(df)
    cont_df.to_csv('Continious_Data_Quality_Report.csv',index=False)
    
# Adding Total Rows Column 
#cont_df['TotalRows'] = len(df)
```
